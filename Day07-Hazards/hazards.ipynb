```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("https://github.com/Rodrigojaure92/30daychartchallenge/raw/refs/heads/main/Day07-Hazards/covid.csv", delimiter = ";")
```

```python
# Configurar el estilo del gr치fico
sns.set(style="whitegrid")

# Crear un gr치fico de densidad de EDAD_DECLARADA diferenciado por SEXO
plt.figure(figsize=(8, 6))
sns.kdeplot(data=data, x="EDAD_DECLARADA", hue="SEXO", fill=True, common_norm=False, alpha=0.5, palette="pastel")
plt.title("Distribuci칩n de densidad de edades por sexo de infectados por COVID-19 (2020-3 al 2024-3)", fontsize=14)
plt.xlabel("Edad Declarada", fontsize=12)
plt.ylabel("Densidad", fontsize=12)

# Mostrar el gr치fico
plt.tight_layout()
plt.show()
```

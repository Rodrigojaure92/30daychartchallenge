```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("https://github.com/Rodrigojaure92/30daychartchallenge/raw/refs/heads/main/Day16-Weather/AirQuality.csv", delimiter=";")
columnas_a_eliminar = ['Unnamed: 15', 'Unnamed: 16']
data = data.drop(columns=columnas_a_eliminar, errors='ignore')
data = data.dropna(axis=0)
```

```python
data['T'] = data['T'].astype(str).str.replace(',', '.').astype(float)
data['PT08.S5(O3)'] = data['PT08.S5(O3)'].astype(float)
data_filtrado = data[(data['T'] >= 0) & (data['PT08.S5(O3)'] >= 0)]

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(data_filtrado['T'], data_filtrado['PT08.S5(O3)'], alpha=0.6, color='teal')
plt.title('Relación entre Temperatura y PT08.S5(O3) - Calidad de aire', fontsize=16)
plt.xlabel('Temperatura (°C)', fontsize=12)
plt.ylabel('PT08.S5(O3) (ug/m^3)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle

data = pd.read_csv("https://github.com/Rodrigojaure92/30daychartchallenge/raw/refs/heads/main/Day04-Waffle/dengue.csv")
```

```python
sexo_counts = {'MASCULINO': 0, 'FEMENINO': 0}

# Recorrer la columna 'SEXO' para contar las ocurrencias manualmente
for sexo in data['SEXO'].dropna():
    if sexo in sexo_counts:
        sexo_counts[sexo] += 1

legend_labels = [f"{sexo}: {count}" for sexo, count in sexo_counts.items()]

fig = plt.figure(
    FigureClass=Waffle,
    values=sexo_counts,
    colors=["#1f77b4", "#ff7f0e"],  # Colores para cada categoría
    title={'label': 'Infectados dengue por Sexo en Perú (ene 2023 a may 2023)', 'fontsize': 14},
    figsize=(12, 6),  # Ajuste del tamaño de la figura
    rows=9,
    columns=9,  # Ajuste el número de columnas
    legend={
        'labels': legend_labels,  # Incluir la cantidad junto al nombre
        'loc': 'upper left',
        'bbox_to_anchor': (1.05, 1)  # Mover la leyenda hacia la derecha
    }
)

# Mostrar el gráfico
plt.show()
```
